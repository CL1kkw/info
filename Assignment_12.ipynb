```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Retail_Sales_Data.csv") 
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sales_by_category = df.groupby('Product Category')['Total Amount'].sum()
print("\nTotal Sales by Product Category:")
print(sales_by_category)
```

```python
sales_by_category.plot(kind='bar')
```

```python
sales_by_category.plot(kind='pie')
```

```python
top_categories = sales_by_category.sort_values(ascending=False)
print("\nTop-Performing Product Categories by Sales Amount:")
print(top_categories)
```

```python
sales_by_gender_category = df.groupby(['Gender', 'Product Category'])['Total Amount'].sum().unstack()
print("\nTotal Sales by Gender and Product Category:")
print(sales_by_gender_category)
```

```python
sales_by_gender_category.plot(kind='bar', stacked=True)
```

```python
sales_by_gender_category.plot(kind='bar', stacked=False)
```
