```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("City_Air_Quality.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
city_name = "Delhi"
df_city = df[df['City'] == city_name]
df_city = df_city.sort_values('Date')
```

```python
plt.plot(df_city['Date'], df_city['AQI'])
```
